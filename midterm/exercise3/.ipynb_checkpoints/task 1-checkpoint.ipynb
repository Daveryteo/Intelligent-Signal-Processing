<h1>Task 1: Extract and read the hidden message </h1>

Import the library and open the file

```python
import wave
import array
sound = wave.open("sounds/Ex3_sound1.wav", "rb")
```

Convert the sound into a byte array

```python
byteArray = array.array("B")
byteArray.fromstring(sound.readframes(sound.getnframes()))
sound.close()
```

Extract the least significant bit

```python
extracted = []

for i in range(len(byteArray)):
    extracted.append(byteArray[i] & 1)
```

<p>Next, we change the byteArray back to a string.</p>
<p>First we loop from 0 to the length of the extracted array and we do it in intervals of 8 bits.</p>
<p>We then use the map function to iterate over the entire byte and convert them all to strings.</p>
<p>We iterate through the map object to concatenate it into an 8bit string</p>
<p>After we obtain the 8bit string, we convert the 8bit string into an integer</p>
<p>Finally we concatenate the result of converting the integer into a character</p>

```python
finalString = ""

for i in range (0, len(extracted), 8):
    firstString = map(str, extracted[i : i + 8])
    secondString = ""
    for char in firstString:
        secondString += char
    thirdString = (int(secondString, 2))
    finalString += chr(thirdString)
```

The final result from the conversion of byteArray to a string

```python
print(finalString)
```

Split the string using # and get the first element from that

```python
hidden_message = finalString.split("#")[0]
print(hidden_message)
```
